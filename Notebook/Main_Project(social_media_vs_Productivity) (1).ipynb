### **6. Data Cleaning & Preprocessing**

Data cleaning and preprocessing is a crucial step in any data analysis project, as it ensures the accuracy, consistency, and reliability of the dataset. In this phase, the raw data is prepared for analysis by handling missing values, removing duplicates, standardizing column names, and creating derived features. These steps help improve data quality and make the dataset suitable for meaningful exploratory analysis.



#### **6.1 Creating a Copy of the Dataset**
 A copy of the original dataset is created to preserve the raw data and ensure that all cleaning and transformation steps are performed on a separate dataset.

```python
clean = df.copy()
```

#### **6.2 Handle missing values**

Missing values in numerical columns are handled using appropriate statistical methods, while categorical columns are treated using the most frequent values. This ensures that the dataset remains complete without significantly affecting data distribution.

```python
# Handle missing values
df = df.fillna(df.median(numeric_only=True))   # numeric → median
df = df.fillna(df.mode().iloc[0])              # categorical → mode
```

#### **6.3 Remove Duplicates**
Duplicate records are identified and removed to prevent repeated observations from biasing the analysis results.

```python
# Remove duplicates
df = df.drop_duplicates()
```

#### **6.3 Converting Data Type**

This code performs data type standardization and value transformation to prepare the dataset for clearer analysis and interpretation.

Categorical Data Conversion: A list of categorical columns (gender, job_type, social_platform_preference) is defined and converted to the category data type. This improves memory efficiency and ensures these variables are treated correctly during analysis and visualization.

Percentage Conversion of Score-Based Columns: Selected numeric columns related to productivity, stress, burnout, and job satisfaction are identified. Each of these columns is multiplied by 100 to convert their values into percentage format, making them more intuitive and comparable.

Rounding Values The transformed percentage values are rounded to two decimal places to improve readability and presentation quality.

Data Preview Finally, df.head() is used to display the updated DataFrame and verify that the transformations were applied correctly.

```python
# changing data_type
cat_cols = ["gender",
"job_type",
"social_platform_preference"
]

for col in cat_cols:
    df[col] = df[col].astype('category')

columns_to_convert = [
    'perceived_productivity_score',
    'actual_productivity_score',
    'job_satisfaction_score',
    'stress_level',
    'days_feeling_burnout_per_month'
]

# Convert these columns to percentage
for col in columns_to_convert:
    df[col] = df[col] * 100

#  Round the values to 2 decimal places
df[columns_to_convert] = df[columns_to_convert].round(2)

# Show updated dataframe
df.head()
```
